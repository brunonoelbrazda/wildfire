# Model Evaluation

```python
c2 = c2[c2['FIRE_SIZE_CLASS']!='G'].copy()
c2 = c2[c2['FIRE_SIZE_CLASS']!='E'].copy()
c2 = c2[c2['FIRE_SIZE_CLASS']!='F'].copy()
```

```python
c2['FIRE_SIZE_CLASS'].value_counts(normalize=True)
```

```python
X = c2[['LATITUDE_x','LONGITUDE_x','STAT_CAUSE_DESCR','DURATION',
       'OWNER_DESCR','COUNTY','STATE','DISCOVERY_DOY','CONT_DOY',
        'HUMIDITY','TEMP','DISCOVERY_MONTH']].copy()
X = pd.get_dummies(X,columns=['STAT_CAUSE_DESCR','OWNER_DESCR','DISCOVERY_MONTH',
                              'DISCOVERY_DOY','CONT_DOY','COUNTY','STATE'],drop_first=True).copy()
scaler = StandardScaler()
X = scaler.fit_transform(X)

y = c2['FIRE_SIZE_CLASS'].copy()

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    stratify=y, 
    test_size=0.2, 
    random_state=1)

model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(cross_val_score(model, X_train, y_train, cv=5).mean())
print(model.score(X_test, y_test))
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
c2_pca = PCA(n_components=5)
c2_pca.fit(X)
```

```python
print('Explained Variance Ratios:')
c2_pca.explained_variance_ratio_
```

```python
c2_pcs = c2_pca.transform(X)
print(np.allclose(np.corrcoef(c2_pcs.T), np.eye(len(c2_pcs.T))))
```

```python
c2_pcs = pd.DataFrame(c2_pca.transform(X), 
                        columns=[f'PC_{i+1}' for i in range(c2_pca.n_components_)])
c2_pcs.head()
```

```python
# Decision Tree Classifier with 3 PCA Components

X = c2_pcs[['PC_1','PC_2','PC_3']]

scaler = StandardScaler()
scaler.fit(X)

y = c2['FIRE_SIZE_CLASS'].copy()

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    stratify=y, 
    test_size=0.2, 
    random_state=1)

model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(cross_val_score(model, X_train, y_train, cv=5).mean())
print(model.score(X_test, y_test))
```

```python
#Random Forest on PCA
X = c2_pcs[['PC_1','PC_2','PC_3']]

scaler = StandardScaler()
scaler.fit(X)

y = c2['FIRE_SIZE_CLASS'].copy()

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    stratify=y, 
    test_size=0.2, 
    random_state=1)

model = RandomForestClassifier(
                           n_estimators=100,
                           random_state=1)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(cross_val_score(model, X_train, y_train, cv=5).mean())
print(model.score(X_test, y_test))
```

```python
#With a mean cv score noticeably higher than baseline, PCA proved to be 
# an affective way of reducing dimensionality. With that being said,
# being unable to extract feature importances made it not particularly 
# useful for acheiving my goal.
pd.DataFrame(model.feature_importances_,
             columns=['importance'],
             index=X.columns
             ).sort_values(by='importance', ascending=False
                           ).plot(kind='barh', figsize=(5, 5))
plt.show()
```

# Random Forest

```python
model = RandomForestClassifier(
                           n_estimators=100,
                           random_state=1)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(cross_val_score(model, X_train, y_train, cv=5).mean())
print(model.score(X_test, y_test))
```

```python
model = RandomForestClassifier(
                           n_estimators=100,
                           random_state=1)
model.fit(X_train, y_train)

print(cross_val_score(model, X_test, y_test, cv=5).mean())
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
plot_confusion_matrix(model, X_test, y_test, cmap='Blues', ax=ax)

plt.show()
```

```python
from sklearn.metrics import classification_report
y_true = y_test
y_pred = model.predict(X_test)

target_names = ['Fire Size A', 'Fire Size B', 'Fire Size C',
               'Fire Size D']
print(classification_report(y_true, y_pred, target_names=target_names))
```

```python
from yellowbrick.classifier import ClassPredictionError

classes = ['A','B','C','D']
visualizer = ClassPredictionError(
    RandomForestClassifier(random_state=1, n_estimators=100), classes=classes
)

# Fit the training data to the visualizer
visualizer.fit(X_train, y_train)

# Evaluate the model on the test data
visualizer.score(X_test, y_test)

# Draw visualization
visualizer.show()
```
